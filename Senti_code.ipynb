```python
from transformers import pipeline
```

```python
data = pd.read_excel('dummy_data.xlsx')
```

```python
data["full_text_len"] = data['text'].apply(lambda x: len(x))
```

```python
pipe = pipeline(task='sentiment-analysis', model='distilbert-base-uncased-finetuned-sst-2-english')
```

```python
def slice_text(text, chunk_size):
    chunks = [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]
    return chunks
```

```python
from collections import Counter
```

```python
def predictions(text):
  chunks = slice_text(text, 512)
  pred_list = [pipe(item)[0].get('label') for item in chunks]
  value_counts = Counter(pred_list)
  max_value = max(value_counts, key=lambda k: value_counts[k])
  return max_value
```

```python
data['predictions'] = data['text'].apply(lambda x: predictions(x))
```
